```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
df = pd.read_csv("Dataset .csv")  # make sure this file is in your working directory

```

```python
# Set seaborn style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

```

```python
# Calculate delivery percentages
delivery_counts = df['Has Online delivery'].value_counts()
delivery_percentage = (delivery_counts / delivery_counts.sum()) * 100

# --- Pie Chart ---
plt.figure(figsize=(8, 6))
colors = ['#66b3ff', '#ff9999']
plt.pie(delivery_percentage, labels=delivery_percentage.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Percentage of Restaurants with Online Delivery")
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()

# --- Bar Chart of Average Ratings ---
avg_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, palette='Set2')
plt.title("Average Ratings: Online Delivery vs No Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")

# Annotate bars
for i, val in enumerate(avg_ratings.values):
    plt.text(i, val + 0.02, f"{val:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
