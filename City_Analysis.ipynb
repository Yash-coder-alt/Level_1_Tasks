```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
df = pd.read_csv("Dataset .csv")  # Ensure this file is in your working directory

```

```python
# Set plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)

```

```python
# 1. City with the highest number of restaurants
city_counts = df['City'].value_counts()
city_with_most_restaurants = city_counts.idxmax()
restaurant_count = city_counts.max()
print(f"City with the Most Restaurants: {city_with_most_restaurants} ({restaurant_count})")
```

```python
# 2. Average rating per city
average_ratings = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
top_rated_city = average_ratings.idxmax()
top_rating_value = average_ratings.max()
print(f"City with the Highest Average Rating: {top_rated_city} ({round(top_rating_value, 2)})")

```

```python
# Plotting
fig, ax = plt.subplots(1, 2, figsize=(20, 7))

# Plot 1: Number of Restaurants per City (Top 10)
sns.barplot(x=city_counts.head(10).values, y=city_counts.head(10).index, ax=ax[0], palette='viridis')
ax[0].set_title("Top 10 Cities by Number of Restaurants")
ax[0].set_xlabel("Number of Restaurants")
ax[0].set_ylabel("City")

# Plot 2: Average Rating per City (Top 10)
sns.barplot(x=average_ratings.head(10).values, y=average_ratings.head(10).index, ax=ax[1], palette='coolwarm')
ax[1].set_title("Top 10 Cities by Average Restaurant Rating")
ax[1].set_xlabel("Average Rating")
ax[1].set_ylabel("City")

plt.tight_layout()
plt.show()
```
