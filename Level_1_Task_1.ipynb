```python
# ðŸ“Œ Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# ðŸ“Œ Step 2: Load the Dataset
# Replace with your actual file path
file_path = 'Dataset .csv'
df = pd.read_csv(file_path)

```

```python
# ðŸ“Œ Step 3: Drop rows with missing 'Cuisines'
df_clean = df.dropna(subset=['Cuisines'])

```

```python
# ðŸ“Œ Step 4: Split and count cuisines
cuisine_counts = df_clean['Cuisines'].str.split(', ').explode().value_counts()

```

```python
# ðŸ“Œ Step 5: Get top 3 most common cuisines
top_3_cuisines = cuisine_counts.head(3)

```

```python
# ðŸ“Œ Step 6: Calculate total restaurants and percentages
total_restaurants = len(df_clean)
top_3_percentages = (top_3_cuisines / total_restaurants * 100).round(2)

```

```python
# ðŸ“Œ Step 7: Create a summary DataFrame
summary_df = pd.DataFrame({
    'Cuisine': top_3_cuisines.index,
    'Count': top_3_cuisines.values,
    'Percentage': top_3_percentages.values
})

```

```python
# Display the summary
print("Top 3 Most Common Cuisines:")
display(summary_df)

```

```python
# ðŸ“Š Step 8: Visualization - Bar Chart
plt.figure(figsize=(8, 5))
bars = plt.bar(summary_df['Cuisine'], summary_df['Percentage'], color='skyblue')
plt.title('Top 3 Most Common Cuisines (by Percentage)')
plt.xlabel('Cuisine')
plt.ylabel('Percentage of Restaurants')
plt.ylim(0, max(summary_df['Percentage']) + 10)

```

```python
# Add percentage labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height}%',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
